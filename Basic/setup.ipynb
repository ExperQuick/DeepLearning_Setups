# ðŸ§ª PyLabFlow Tutorial: Creating Your Lab (Project Workspace)

Welcome to your first PyLabFlow setup!  
Every experiment in PyLabFlow runs **inside a lab**, which acts as your projectâ€™s central workspace.

A *lab* automatically manages:

- ðŸ“ A project directory (where all experiment data lives)
- ðŸ§© A component directory (for your datasets, models, and transforms)
- ðŸ—‚ï¸ Databases for tracking logs and workflows
- âš™ï¸ A JSON settings file to easily reload this environment later

Once created, you can **resume, clone, or share** your lab at any time.

Letâ€™s start by creating your first one ðŸ‘‡


`````python
# Step 1ï¸âƒ£: Import the PyLabFlow lab creation utility
from plf.lab import create_project

# Step 2ï¸âƒ£: Define your lab configuration

settings = {
    "project_name": "PLF_Starter",   # A name for your project workspace
    "project_dir": "./",             # Root directory where lab data will live
    "component_dir": "StarterCompDir" # Directory for user components (models, datasets, etc.)
}

# ðŸ’¡ Note:
# The component directory can be an existing folder or new â€” 
# PyLabFlow will handle it automatically.
`````

`````python
# Step 3ï¸âƒ£: Create your PyLabFlow project (Lab)
settings_path = create_project(settings)

# Print the absolute path to your lab settings file
print("âœ… Lab created successfully!")
print("Settings file path:", settings_path)
`````

### ðŸ§  What `create_project()` Did

PyLabFlow just initialized a **complete workspace** for you:

| Item | Description |
|------|--------------|
| `PLF_Starter/` | Main lab folder |
| `PLF_Starter/logs.db` | Stores experiment call logs |
| `PLF_Starter/ppls.db` | Tracks workflows (pipelines) and their status |
| `PLF_Starter/Archived/ppls.db` | Keeps old or frozen pipeline states |
| `StarterCompDir/` |  will adds it to python module path so that all files/components re directly importable whene the this lab is used
| `PLF_Starter.json` | JSON configuration file for reloading the lab later |

Every time you create a workflow or run an experiment,  
PyLabFlow records metadata and results into these structured databases.

## âš™ï¸ Next Step: Activate Your Lab

Now that your lab is created, youâ€™ll use the generated `.json` file 
to re-enter it in future sessions or scripts.

Call `lab_setup()` with your settings file path before running workflows:


`````python
from plf.lab import lab_setup

# Pass the same path printed above
lab_setup(settings_path)

print("ðŸ”§ Lab environment loaded successfully!")

# check if lab is ready

from plf.experiment import PipeLine

P = PipeLine()
print(P.settings) # it will print the same dictionary given in create_project
`````
